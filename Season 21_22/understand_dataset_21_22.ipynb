```python
# Rename columns for better readability
rank_outcomes_data.columns = [
    'Index', 'Teams Header', 'Rank', 'P Header', 'Team', 'NaN1', 'Matches', 
    'Wins', 'Draws', 'Losses', 'Goals Scored', 'Goals Conceded', 
    'Goal Difference', 'Points'
]

# Drop the first two rows which contain header information and NaN values
rank_outcomes_data = rank_outcomes_data.drop([0, 1])

# Drop unnecessary columns
rank_outcomes_data = rank_outcomes_data.drop(columns=['Index', 'Teams Header', 'P Header', 'NaN1'])

# Reset the index
rank_outcomes_data.reset_index(drop=True, inplace=True)



# Display the cleaned data
rank_outcomes_data
```

Normalize Team Names

```python
# Create a mapping dictionary to standardize team names
team_name_mapping = {
    'TRABZONSPOR': 'Trabzonspor',
    'FENERBAHÇE': 'Fenerbahce',
    'KONYASPOR': 'Konyaspor',
    'BASAKSEHIR': 'Istanbul Basaksehir',
    'ALANYASPOR': 'Alanyaspor',
    'BESIKTAS': 'Besiktas',
    'ANTALYASPOR': 'Antalyaspor',
    'FATIH KARAGÜMRÜK': 'Fatih Karagumruk',
    'ADANA DEMIRSPOR': 'Adana Demirspor',
    'SIVASSPOR': 'Sivasspor',
    'KASIMPASA': 'Kasimpasa',
    'HATAYSPOR': 'Hatayspor',
    'GALATASARAY': 'Galatasaray',
    'KAYSERISPOR': 'Kayserispor',
    'GAZIANTEP FK': 'Gaziantep FK',
    'GIRESUNSPOR': 'Giresunspor',
    'ÇAYKUR RIZESPOR': 'Rizespor',
    'ALTAY': 'Altay',
    'GÖZTEPE': 'Goztepe',
    'YENI MALATYASPOR': 'Yeni Malatyaspor'
}


# Apply the mapping to the rank_outcomes_data
rank_outcomes_data['Team'] = rank_outcomes_data['Team'].map(team_name_mapping)

# Verify the updated team names in rank_outcomes_data
rank_outcomes_data['Team'].unique()
```

Merging Dataset

```python
# Merge the teams_data with rank_outcomes_data based on the Team column
merged_data = pd.merge(teams_data, rank_outcomes_data, on='Team', how='inner')

# Display the merged data
merged_data.to_excel('merged_data_21_22.xlsx')
```

```python
merged_data = merged_data[~merged_data['Match'].str.contains("Ortalama|Toplam")] # we don't need it anymore

def convert_to_minutes(time_str):
    if pd.isnull(time_str) or time_str == "":
        return None
    # Ensure the time string is in string format
    time_str = str(time_str)
    # Convert to timedelta and then to total seconds, and finally to minutes
    td = pd.to_timedelta(time_str)
    return td.total_seconds() / 60



# Apply the conversion function to your time columns
merged_data['Minutes on the field total attacking'] = merged_data['Minutes on the field total attacking'].apply(convert_to_minutes)
merged_data['Minutes on the field total defending'] = merged_data['Minutes on the field total defending'].apply(convert_to_minutes)
merged_data['Min ball in play'] = merged_data['Min ball in play'].apply(convert_to_minutes)

# Check
merged_data[['Minutes on the field total attacking', 'Minutes on the field total defending', 'Min ball in play']].head()
```
