```python
import pandas as pd

# Load the Excel file to check its contents
file_path = '22_23_season.xlsx'
teams_data = pd.read_excel(file_path, sheet_name='Teams')
teams_data
```

```python
teams_data.columns = [
    'Team', 'Match', 'Match Date', 'Total Distance Covered', 
    'Distance Covered Speed 0-0.2 m/s', 'Distance Covered Speed 0.21-2 m/s', 
    'Distance Covered Speed 2.01-4 m/s', 'Distance Covered Speed 4.01-5.5 m/s', 
    'Distance Covered Speed 5.51-7 m/s', 'Distance Covered Speed >7 m/s', 
    'No. of Accelerations 5.5-7 m/s²','No. of Accelerations >7 m/s²',
    'Maximal Speed', 'Minutes on the field total attacking', 
    'Minutes on the field total defending', 'Min ball in play',
]
```

```python
import re

def extract_scores(match):
    try:
        # Match the pattern with a regular expression
        match_pattern = re.match(r'(.+?) (\d+:\d+) (.+)', match)
        home_team = match_pattern.group(1)
        score = match_pattern.group(2)
        away_team = match_pattern.group(3)
        home_score, away_score = map(int, score.split(':'))
        return home_team, home_score, away_team, away_score
    except:
        return None, None, None, None

# Apply the function to extract scores
teams_data[['Home Team', 'Home Score', 'Away Team', 'Away Score']] = teams_data['Match'].apply(lambda x: pd.Series(extract_scores(x)))

# Display the updated DataFrame
teams_data
```

```python
data = teams_data
cumulative_columns = [
    'Cumulative Points', 'Cumulative Goal Difference', 'Cumulative Goals Scored',
    'Cumulative Goals Conceded', 'Cumulative Wins', 'Cumulative Losses',
    'Cumulative Ties', 'Current Form Points', 'Current Form Goal Difference',
    'Current Form Goals Scored', 'Current Form Goals Conceded', 'Current Form Wins',
    'Current Form Losses', 'Current Form Ties'
]
for col in cumulative_columns:
    data[col] = 0

# Initialize a dictionary to hold cumulative stats for each team
cumulative_stats = {}
current_team = None

# Iterate through each row in the DataFrame
for index, row in data.iterrows():
    team = row['Team']
    home_team = row['Home Team']
    away_team = row['Away Team']
    home_score = row['Home Score']
    away_score = row['Away Score']
    
    if team not in cumulative_stats:
        # Reset cumulative stats for a new team
        cumulative_stats[team] = {
            'Points': 0, 'Goal Difference': 0, 'Goals Scored': 0, 'Goals Conceded': 0,
            'Wins': 0, 'Losses': 0, 'Ties': 0,
            'Form Points': [], 'Form Goal Difference': [], 'Form Goals Scored': [],
            'Form Goals Conceded': [], 'Form Wins': [], 'Form Losses': [], 'Form Ties': []
        }

    # Initialize temporary variables for the current row's update
    match_points = 0
    match_goal_diff = 0
    match_goals_scored = 0
    match_goals_conceded = 0
    match_win = 0
    match_loss = 0
    match_tie = 0

    if home_team == team:
        # Calculate match stats for home team
        match_goals_scored = home_score
        match_goals_conceded = away_score
        match_goal_diff = home_score - away_score

        if home_score > away_score:
            match_points = 3
            match_win = 1
        elif home_score < away_score:
            match_points = 0
            match_loss = 1
        else:
            match_points = 1
            match_tie = 1
    elif away_team == team:
        # Calculate match stats for away team
        match_goals_scored = away_score
        match_goals_conceded = home_score
        match_goal_diff = away_score - home_score

        if away_score > home_score:
            match_points = 3
            match_win = 1
        elif away_score < home_score:
            match_points = 0
            match_loss = 1
        else:
            match_points = 1
            match_tie = 1

    # Update the DataFrame with cumulative stats
    data.at[index, 'Cumulative Points'] = cumulative_stats[team]['Points']
    data.at[index, 'Cumulative Goal Difference'] = cumulative_stats[team]['Goal Difference']
    data.at[index, 'Cumulative Goals Scored'] = cumulative_stats[team]['Goals Scored']
    data.at[index, 'Cumulative Goals Conceded'] = cumulative_stats[team]['Goals Conceded']
    data.at[index, 'Cumulative Wins'] = cumulative_stats[team]['Wins']
    data.at[index, 'Cumulative Losses'] = cumulative_stats[team]['Losses']
    data.at[index, 'Cumulative Ties'] = cumulative_stats[team]['Ties']

    # Update current form stats (last 5 matches)
    data.at[index, 'Current Form Points'] = sum(cumulative_stats[team]['Form Points'][-5:])
    data.at[index, 'Current Form Goal Difference'] = sum(cumulative_stats[team]['Form Goal Difference'][-5:])
    data.at[index, 'Current Form Goals Scored'] = sum(cumulative_stats[team]['Form Goals Scored'][-5:])
    data.at[index, 'Current Form Goals Conceded'] = sum(cumulative_stats[team]['Form Goals Conceded'][-5:])
    data.at[index, 'Current Form Wins'] = sum(cumulative_stats[team]['Form Wins'][-5:])
    data.at[index, 'Current Form Losses'] = sum(cumulative_stats[team]['Form Losses'][-5:])
    data.at[index, 'Current Form Ties'] = sum(cumulative_stats[team]['Form Ties'][-5:])

    # Update the cumulative stats after processing the current match
    cumulative_stats[team]['Points'] += match_points
    cumulative_stats[team]['Goal Difference'] += match_goal_diff
    cumulative_stats[team]['Goals Scored'] += match_goals_scored
    cumulative_stats[team]['Goals Conceded'] += match_goals_conceded
    cumulative_stats[team]['Wins'] += match_win
    cumulative_stats[team]['Losses'] += match_loss
    cumulative_stats[team]['Ties'] += match_tie

    # Append current match stats to form lists
    cumulative_stats[team]['Form Points'].append(match_points)
    cumulative_stats[team]['Form Goal Difference'].append(match_goal_diff)
    cumulative_stats[team]['Form Goals Scored'].append(match_goals_scored)
    cumulative_stats[team]['Form Goals Conceded'].append(match_goals_conceded)
    cumulative_stats[team]['Form Wins'].append(match_win)
    cumulative_stats[team]['Form Losses'].append(match_loss)
    cumulative_stats[team]['Form Ties'].append(match_tie)
```

```python
players_data = pd.read_excel(file_path, sheet_name='Players')
players_data.columns = [
    'Last Name', 'Match', 'Match Date', 'Team', 'Minutes Played', 
    'Total Distance Covered', 'Distance Covered Speed 0-0.2 m/s', 
    'Distance Covered Speed 0.21-2 m/s', 'Distance Covered Speed 2.01-4 m/s', 
    'Distance Covered Speed 4.01-5.5 m/s', 'Distance Covered Speed 5.51-7 m/s', 
    'Distance Covered Speed >7 m/s', 'No. of Accelerations 5.5-7 m/s²', 
    'No. of Accelerations >7 m/s²', 'Maximal Speed', 
    'High Speed Runs Attacking', 'High Speed Runs Defensive', 
    'Sprints Attacking', 'Sprints Defensive'
]
players_data
```

```python
additional_data = players_data.drop(columns=['Last Name','Minutes Played', 
    'Total Distance Covered', 'Distance Covered Speed 0-0.2 m/s', 
    'Distance Covered Speed 0.21-2 m/s', 'Distance Covered Speed 2.01-4 m/s', 
    'Distance Covered Speed 4.01-5.5 m/s', 'Distance Covered Speed 5.51-7 m/s', 
    'Distance Covered Speed >7 m/s', 'No. of Accelerations 5.5-7 m/s²', 
    'No. of Accelerations >7 m/s²', 'Maximal Speed'])

additional_data_aggregated = additional_data.groupby(['Match', 'Team', 'Match Date']).sum().reset_index()
# Merge the aggregated additional data with the main dataset on 'Match', 'Team', and 'Match Date'
merged_data = pd.merge(data, additional_data_aggregated, on=['Match', 'Team', 'Match Date'], how='left')

merged_data
```

```python
# Add a column to indicate if the team is playing at home or away
merged_data['Home/Away'] = merged_data.apply(lambda row: 'Home' if row['Home Team'] == row['Team'] else 'Away', axis=1)
merged_data['Goals Scored'] = merged_data.apply(lambda row: row['Home Score'] if row['Home Team'] == row['Team'] else row['Away Score'], axis=1)
merged_data['Goals Conceded'] = merged_data.apply(lambda row: row['Away Score'] if row['Home Team'] == row['Team'] else row['Home Score'], axis=1)

# Calculate the goal difference for the specific match
merged_data['Goal Difference'] = merged_data['Goals Scored'] - merged_data['Goals Conceded']
```

```python
def determine_result(goal_difference):
    if goal_difference > 0:
        return 'Win'
    elif goal_difference < 0:
        return 'Lose'
    else:
        return 'Tie'

# Apply the function to create the target variable
merged_data['Result'] = merged_data['Goal Difference'].apply(determine_result)

merged_data
```

```python
merged_data = merged_data[~merged_data['Match'].str.contains("Ortalama|Toplam")] # we don't need it anymore

merged_data['season'] = '2023'
promoted_teams = ['MKE Ankaragucu', 'Istanbulspor', 'Umraniyespor']

# Add the 'Promoted_from_19_20_home' column
merged_data['Promoted_from_21_22_home'] = merged_data['Home Team'].apply(lambda x: 1 if x in promoted_teams else 0)

# Add the 'Promoted_from_19_20_away' column
merged_data['Promoted_from_21_22_away'] = merged_data['Away Team'].apply(lambda x: 1 if x in promoted_teams else 0)
```
