```python
import pandas as pd

# Load the Excel file to check its contents
file_path = '20_21_season.xlsx'
teams_data = pd.read_excel(file_path, sheet_name='Teams')
teams_data
```

```python
teams_data.columns = [
    'Team', 'Match', 'Match Date', 'Total Distance Covered', 
    'Distance Covered Speed 0-0.2 m/s', 'Distance Covered Speed 0.21-2 m/s', 
    'Distance Covered Speed 2.01-4 m/s', 'Distance Covered Speed 4.01-5.5 m/s', 
    'Distance Covered Speed 5.51-7 m/s', 'Distance Covered Speed >7 m/s', 
    'No. of Accelerations 5.5-7 m/s²', 'No. of Accelerations >7 m/s²', 
    'Maximal Speed'
]
```

```python
import re

def extract_scores(match):
    try:
        # Match the pattern with a regular expression
        match_pattern = re.match(r'(.+?) (\d+:\d+) (.+)', match)
        home_team = match_pattern.group(1)
        score = match_pattern.group(2)
        away_team = match_pattern.group(3)
        home_score, away_score = map(int, score.split(':'))
        return home_team, home_score, away_team, away_score
    except:
        return None, None, None, None

# Apply the function to extract scores
teams_data[['Home Team', 'Home Score', 'Away Team', 'Away Score']] = teams_data['Match'].apply(lambda x: pd.Series(extract_scores(x)))

# Display the updated DataFrame
teams_data
```

```python
data = teams_data
import pandas as pd
data['Cumulative Points'] = 0
data['Cumulative Goal Difference'] = 0
data['Cumulative Goals Scored'] = 0
data['Cumulative Goals Conceded'] = 0
data['Cumulative Wins'] = 0
data['Cumulative Losses'] = 0
data['Cumulative Ties'] = 0
data['Current Form Points'] = 0
data['Current Form Goal Difference'] = 0
data['Current Form Goals Scored'] = 0
data['Current Form Goals Conceded'] = 0
data['Current Form Wins'] = 0
data['Current Form Losses'] = 0
data['Current Form Ties'] = 0

# Initialize a dictionary to hold cumulative stats for each team
cumulative_stats = {}
current_team = None

for index, row in data.iterrows():
    team = row['Team']
    if team != current_team:
        # Reset cumulative stats for a new team
        current_team = team
        cumulative_stats[team] = {
            'Points': 0, 'Goal Difference': 0, 'Goals Scored': 0, 'Goals Conceded': 0,
            'Wins': 0, 'Losses': 0, 'Ties': 0
        }
        # Create a temporary DataFrame to hold the last 5 matches for calculating current form
        temp_df = pd.DataFrame(columns=['Points', 'Goal Difference', 'Goals Scored', 'Goals Conceded', 'Wins', 'Losses', 'Ties'])

    home_team = row['Home Team']
    away_team = row['Away Team']
    home_score = row['Home Score']
    away_score = row['Away Score']
    
    if home_team == team:
        # Update stats for the home team
        if home_score > away_score:
            cumulative_stats[home_team]['Wins'] += 1
            cumulative_stats[home_team]['Points'] += 3
            match_points = 3
        elif home_score < away_score:
            cumulative_stats[home_team]['Losses'] += 1
            match_points = 0
        else:
            cumulative_stats[home_team]['Ties'] += 1
            cumulative_stats[home_team]['Points'] += 1
            match_points = 1

        cumulative_stats[home_team]['Goals Scored'] += home_score
        cumulative_stats[home_team]['Goals Conceded'] += away_score
        cumulative_stats[home_team]['Goal Difference'] = cumulative_stats[home_team]['Goals Scored'] - cumulative_stats[home_team]['Goals Conceded']

        # Update the temporary DataFrame
        temp_row = pd.DataFrame([{
            'Points': match_points,
            'Goal Difference': home_score - away_score,
            'Goals Scored': home_score,
            'Goals Conceded': away_score,
            'Wins': 1 if match_points == 3 else 0,
            'Losses': 1 if match_points == 0 else 0,
            'Ties': 1 if match_points == 1 else 0
        }])
        temp_df = pd.concat([temp_df, temp_row], ignore_index=True)

        # Calculate rolling sum for the last 5 matches
        if len(temp_df) > 5:
            temp_df = temp_df.iloc[1:]

        current_form = temp_df.sum()

        # Update the dataframe
        data.at[index, 'Cumulative Points'] = cumulative_stats[home_team]['Points']
        data.at[index, 'Cumulative Goal Difference'] = cumulative_stats[home_team]['Goal Difference']
        data.at[index, 'Cumulative Goals Scored'] = cumulative_stats[home_team]['Goals Scored']
        data.at[index, 'Cumulative Goals Conceded'] = cumulative_stats[home_team]['Goals Conceded']
        data.at[index, 'Cumulative Wins'] = cumulative_stats[home_team]['Wins']
        data.at[index, 'Cumulative Losses'] = cumulative_stats[home_team]['Losses']
        data.at[index, 'Cumulative Ties'] = cumulative_stats[home_team]['Ties']
        data.at[index, 'Current Form Points'] = current_form['Points']
        data.at[index, 'Current Form Goal Difference'] = current_form['Goal Difference']
        data.at[index, 'Current Form Goals Scored'] = current_form['Goals Scored']
        data.at[index, 'Current Form Goals Conceded'] = current_form['Goals Conceded']
        data.at[index, 'Current Form Wins'] = current_form['Wins']
        data.at[index, 'Current Form Losses'] = current_form['Losses']
        data.at[index, 'Current Form Ties'] = current_form['Ties']
        
    elif away_team == team:
        # Update stats for the away team
        if away_score > home_score:
            cumulative_stats[away_team]['Wins'] += 1
            cumulative_stats[away_team]['Points'] += 3
            match_points = 3
        elif away_score < home_score:
            cumulative_stats[away_team]['Losses'] += 1
            match_points = 0
        else:
            cumulative_stats[away_team]['Ties'] += 1
            cumulative_stats[away_team]['Points'] += 1
            match_points = 1

        cumulative_stats[away_team]['Goals Scored'] += away_score
        cumulative_stats[away_team]['Goals Conceded'] += home_score
        cumulative_stats[away_team]['Goal Difference'] = cumulative_stats[away_team]['Goals Scored'] - cumulative_stats[away_team]['Goals Conceded']

        # Update the temporary DataFrame
        temp_row = pd.DataFrame([{
            'Points': match_points,
            'Goal Difference': away_score - home_score,
            'Goals Scored': away_score,
            'Goals Conceded': home_score,
            'Wins': 1 if match_points == 3 else 0,
            'Losses': 1 if match_points == 0 else 0,
            'Ties': 1 if match_points == 1 else 0
        }])
        temp_df = pd.concat([temp_df, temp_row], ignore_index=True)

        # Calculate rolling sum for the last 5 matches
        if len(temp_df) > 5:
            temp_df = temp_df.iloc[1:]

        current_form = temp_df.sum()

        # Update the dataframe
        data.at[index, 'Cumulative Points'] = cumulative_stats[away_team]['Points']
        data.at[index, 'Cumulative Goal Difference'] = cumulative_stats[away_team]['Goal Difference']
        data.at[index, 'Cumulative Goals Scored'] = cumulative_stats[away_team]['Goals Scored']
        data.at[index, 'Cumulative Goals Conceded'] = cumulative_stats[away_team]['Goals Conceded']
        data.at[index, 'Cumulative Wins'] = cumulative_stats[away_team]['Wins']
        data.at[index, 'Cumulative Losses'] = cumulative_stats[away_team]['Losses']
        data.at[index, 'Cumulative Ties'] = cumulative_stats[away_team]['Ties']
        data.at[index, 'Current Form Points'] = current_form['Points']
        data.at[index, 'Current Form Goal Difference'] = current_form['Goal Difference']
        data.at[index, 'Current Form Goals Scored'] = current_form['Goals Scored']
        data.at[index, 'Current Form Goals Conceded'] = current_form['Goals Conceded']
        data.at[index, 'Current Form Wins'] = current_form['Wins']
        data.at[index, 'Current Form Losses'] = current_form['Losses']
        data.at[index, 'Current Form Ties'] = current_form['Ties']
data
```

```python
players_data = pd.read_excel(file_path, sheet_name='Players')
players_data.columns = [
    'Last Name', 'Match', 'Match Date', 'Team', 'Minutes Played', 
    'Total Distance Covered', 'Distance Covered Speed 0-0.2 m/s', 
    'Distance Covered Speed 0.21-2 m/s', 'Distance Covered Speed 2.01-4 m/s', 
    'Distance Covered Speed 4.01-5.5 m/s', 'Distance Covered Speed 5.51-7 m/s', 
    'Distance Covered Speed >7 m/s', 'No. of Accelerations 5.5-7 m/s²', 
    'No. of Accelerations >7 m/s²', 'Maximal Speed', 
    'High Speed Runs Attacking', 'High Speed Runs Defensive', 
    'Sprints Attacking', 'Sprints Defensive'
]
players_data
```

```python
additional_data = players_data.drop(columns=['Last Name','Minutes Played', 
    'Total Distance Covered', 'Distance Covered Speed 0-0.2 m/s', 
    'Distance Covered Speed 0.21-2 m/s', 'Distance Covered Speed 2.01-4 m/s', 
    'Distance Covered Speed 4.01-5.5 m/s', 'Distance Covered Speed 5.51-7 m/s', 
    'Distance Covered Speed >7 m/s', 'No. of Accelerations 5.5-7 m/s²', 
    'No. of Accelerations >7 m/s²', 'Maximal Speed'])

additional_data_aggregated = additional_data.groupby(['Match', 'Team', 'Match Date']).sum().reset_index()
# Merge the aggregated additional data with the main dataset on 'Match', 'Team', and 'Match Date'
merged_data = pd.merge(data, additional_data_aggregated, on=['Match', 'Team', 'Match Date'], how='left')

merged_data
```

```python
# Add a column to indicate if the team is playing at home or away
merged_data['Home/Away'] = merged_data.apply(lambda row: 'Home' if row['Home Team'] == row['Team'] else 'Away', axis=1)

# Calculate the goal difference for the specific match
merged_data['Goal Difference'] = merged_data['Home Score'] - merged_data['Away Score']
```

```python
def determine_result(goal_difference):
    if goal_difference > 0:
        return 'Win'
    elif goal_difference < 0:
        return 'Lose'
    else:
        return 'Tie'

# Apply the function to create the target variable
merged_data['Result'] = merged_data['Goal Difference'].apply(determine_result)

merged_data
```
